```python
# Sequential search
def sequential_search(source, key):
    index = 0
    source_len = len(source)
    while index < source_len:
        if key == source[index]:
            return index
        index += 1
    return -1


def sequential_search2(source, key):
    for index in range(len(source)):
        if key ==source[index]:
            return index
    return -1


def sequential_search3(source, key):
    for index, char in enumerate(source):
        if char == key:
            return index
    return -1


source = 'Lux, the Lady of Luminosity'
r1 = sequential_search(source, 'x')
r2 = sequential_search2(source, 'x')
r3 = sequential_search3(source, 'x')
print(r1, r2, r3)
```
